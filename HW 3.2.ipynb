# Limits and Continuity

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    """
    Args:
    x (float): Input value.

    Returns:
    float: Value of f(x).
    """
    return np.sin(x) / x if x != 0 else 1

# Generate random values around zero
np.random.seed(0)  # For reproducibility
random_values = np.random.uniform(-0.1, 0.1, 1000)  
y_values = np.array([f(x) for x in random_values])  

# Plotting limits and continuity
plt.figure(figsize=(10, 6))
plt.scatter(random_values, y_values, alpha=0.5, label=r'$f(x) = \frac{\sin(x)}{x}$', color='blue')
plt.axhline(0, color='black', linewidth=0.5, ls='--') 
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.title(r'Limits and Continuity of $f(x) = \frac{\sin(x)}{x}$ (Random Variables)')  
plt.xlim(-0.1, 0.1)  
plt.ylim(0, 1.5) 
plt.xlabel('x') 
plt.ylabel('f(x)')  

```

# Derivatives

```python
def f_derivative(x):
    """
    Args:
    x (float): Point at which to evaluate the derivative.
    h (float): Small step size for numerical differentiation.
    Returns:
    float: Approximate derivative of f(x) at point x.
    """
    return 2 * x

# Generate random values
random_x = np.random.uniform(-10, 10, 100)
y_values = random_x**2
y_derivative = f_derivative(random_x)

# Plotting the function and its derivative
plt.figure(figsize=(10, 6))
plt.scatter(random_x, y_values, label=r'$f(x) = x^2$', color='blue', alpha=0.6)
plt.scatter(random_x, y_derivative, label=r"$f'(x) = 2x$", color='red', alpha=0.6, marker='x')
plt.title('Function and Its Derivative (Random Variables)')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.xlim(-10, 10)
plt.ylim(-10, 100)
plt.xlabel('x')
plt.ylabel('f(x), f\'(x)')
```
