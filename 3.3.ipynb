```python
import numpy as np
import matplotlib.pyplot as plt
```

# Necessary and Sufficient Conditions of Local Minimzers

```python
def f(x):
    return x**3 - 3*x**2 + 4
def f_prime(x):
    return 3*x**2 - 6*x
def f_double_prime(x):
    return 6*x - 6

# Find critical points 
critical_points = np.roots([3, -6, 0])  # Solving 3x^2 - 6x = 0
# Second derivative to classify the critical points
for point in critical_points:
    second_deriv_value = f_double_prime(point)
    if second_deriv_value > 0:
        print(f"Point {point} is a local minimum.")
    elif second_deriv_value < 0:
        print(f"Point {point} is a local maximum.")
    else:
        print(f"Point {point} is a saddle point.")

x_vals = np.linspace(-1, 3, 400)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label="f(x) = x^3 - 3x^2 + 4")
plt.scatter(critical_points, f(critical_points), color='red', zorder=5, label="Critical Points")
plt.title("Necessary and Sufficient Conditions for Local Minimizers")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

# Convexity and Global Minimizers

```python
def f(x):
    return x**2 + 2*x + 1
# First derivative (gradient) of the function
def f_prime(x):
    return 2 * x + 2
# Find the global minimizer by solving f'(x) = 0
global_minimizer = -1  # Solving 2x + 2 = 0, x = -1
# Plot the convex function
x_vals = np.linspace(-3, 1, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x^2 + 2x + 1")
plt.scatter([global_minimizer], [f(global_minimizer)], color='red', zorder=5, label="Global Minimizer")
plt.title("Convexity and Global Minimizers")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
```

# Gradient Descent

```python
def f(x):
    return x**2 + 2*x + 1
def f_prime(x):
    return 2*x + 2
# Gradient Descent function using only NumPy
def gradient_descent(f_prime, x_init, learning_rate, n_iter):
    x = x_init
    x_values = [x]
    for i in range(n_iter):
        grad = f_prime(x)
        x = x - learning_rate * grad
        x_values.append(x)
    return np.array(x_values)

# Parameters for gradient descent
x_init = 0  # Initial guess
learning_rate = 0.1  # Step size
n_iter = 50  # Number of iterations

# Running Gradient Descent
x_values = gradient_descent(f_prime, x_init, learning_rate, n_iter)

# Plotting the descent process
x_vals = np.linspace(-3, 1, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x^2 + 2x + 1")
plt.scatter(x_values, f(x_values), color='red', zorder=5, label="Gradient Descent Path")
plt.title("Gradient Descent Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
```
