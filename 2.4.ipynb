```python
import numpy as np
import matplotlib.pyplot as plt
```

# MLE for Random Sampels

```python
n = 100
data = np.random.normal(loc = 5, scale = 2, size = n)
sigma  = 1
mu = 0
learn = 0.01
for _ in range(1000):
    mu_gradient = np.sum((data - mu) / (sigma**2))
    sigma_gradient = np.sum(((data - mu)**2) / (sigma**3)) - n / sigma

    mu += learn * mu_gradient
    sigma += learn * sigma_gradient
print(f"Estimated Mean : ", mu)
print("Estimated Standard Deviation: ", sigma)
```

# Linear Regression

```python
x = 2 * np.random.randn(n, 1)
y = 4 + 3 * x + np.random.randn(n, 1)
x_b = np.c_[np.ones((n, 1)), x]
best_theta = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
y_pred = x_b.dot(best_theta)

plt.scatter(x, y, color = "red", label = "Data points")
plt.plot(x, y_pred, label = "Regression line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

print("Estimated Intercept: ", best_theta)
print("Estimated Slope (Coefficient): ", best_theta)
```
